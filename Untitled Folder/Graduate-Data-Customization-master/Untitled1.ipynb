```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
xs = np.arange(1,10)
ys = np.log(xs)
plt.plot(xs,ys,'go')
```

```python
x = np.random.randn(8873)
y = np.random.randn(8873)

heatmap,xedges,yedges = np.histogram2d(x,y,bins=50)
extent = [xedges[0],xedges[-1],yedges[0],yedges[-1]]
plt.imshow(heatmap,extent=extent)
```
