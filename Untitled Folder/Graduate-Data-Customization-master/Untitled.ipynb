```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
recent_grads = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv')
```

```python
##Data Cleaning 
#We are first inspecting the data to see there are NaN values 
print("original df row numbers:",recent_grads.shape[0])
#Now,we will create a new df with all NaN's removed,and we will compare the 
#number of rows in each df 
recent_grads = recent_grads.dropna([0])
print("Cleaned df row numbers:",recent_grads.shape[0])
```
